```python
'''Q1.How can you create a Bokeh plot using Python code?

Ans-To create a Bokeh plot using Python code, you need to follow these steps:
1)Install Bokeh: If you haven't installed Bokeh yet, you can use pip or conda to install it. 
Run either of the following commands in your terminal or command prompt:pip install bokeh
2)Import the necessary modules: In your Python code, import the required modules from the Bokeh
library. Typically, you will need to import bokeh.plotting to create a plot and bokeh.io to 
show or save the plot.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
3)Prepare your data: Set up the data that you want to plot. Ensure that your data is in the 
appropriate format, such as lists or NumPy arrays, depending on your requirements.
4)Create a figure: Initialize a figure object, which will serve as the canvas for your plot.
You can customize various aspects of the plot, such as the title, axes labels, and plot 
dimensions, by passing the corresponding arguments to the figure function.
5)Add glyphs: Use the p.circle(), p.line(), or other glyph methods to add visual elements to
your plot. These glyphs represent data points or lines on the plot. Customize the appearance of
the glyphs by passing additional arguments to these methods.
6)Display the plot: Use output_notebook() to configure Bokeh to display the plot in a Jupyter
Notebook or output_file() to save the plot as an HTML file. Finally, call show(p) to display
the plot.

Q2.What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Ans-
In Bokeh, glyphs are visual markers or shapes that represent data points on a plot. They can be
used to display various types of data, such as scatter plots, lines, bars, and more. Glyphs
allow you to visualize your data in a flexible and customizable manner
To add glyphs to a Bokeh plot, you need to use the corresponding glyph methods provided by the 
figure object. Some commonly used glyph methods include circle(), line(), rect(), patch(), and
image().
Example
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure
p = figure(title="Glyph Example", x_axis_label="X-axis", y_axis_label="Y-axis", plot_width=400, plot_height=300)

# Add glyphs
p.circle(x, y, size=10, color="red", alpha=0.5)  # Adds circles with customized size, color, and transparency
p.line(x, y, line_width=2, color="blue")  # Adds a line connecting the data points

# Display the plot
output_notebook()  # For Jupyter Notebook
show(p)

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

Ans-
In Bokeh, you can customize the appearance of a plot by modifying various aspects such as the 
axes, title, legend, colors, fonts, and more. Here are some common customization options:

1)Axes customization:
Set axis labels: Use the x_axis_label and y_axis_label attributes of the figure object to set
the labels for the x-axis and y-axis, respectively.Adjust axis limits: Use the x_range and 
y_range attributes to set the desired range for the x-axis and y-axis, respectively.Modify tick 
abels: Customize the appearance of tick labels using the xaxis.major_label_text_font_size and 
yaxis.major_label_text_font_size attributes, among others.

2)Title customization:
Set the plot title: Use the title attribute of the figure object to set the plot's title.
Customize title properties: Adjust the title's font size, color, alignment, and other
attributes using the title.text_font_size, title.text_color, title.text_align, and similar 
attributes.

3)Legend customization:
Add a legend: Use the legend attribute of the glyph methods to specify the label for the glyph
and set legend_label to provide a name for the legend.Customize legend properties: Adjust the
font size, color, background color, and other attributes of the legend using the 
legend.label_text_font_size, legend.label_text_color, legend.background_fill_color, and similar
attributes.

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be 
updated in real time?

Ans-
The Bokeh server is a key feature of the Bokeh library that allows you to create interactive 
plots that can be updated and respond to user interactions in real time. The Bokeh server is a 
Python process that runs separately and communicates with the browser-based Bokeh client using 
websockets.By leveraging the Bokeh server, you can create dynamic and interactive visualizations that can 
update based on user inputs, streaming data, or periodic updates. This enables the creation of 
live dashboards, real-time monitoring systems, or any application where the plot needs to update
dynamically.
To use the Bokeh server to create interactive plots that update in real time, follow these steps:
1)Install Bokeh: Ensure that you have Bokeh installed. You can use pip or conda to install it.
2)Import necessary modules: In your Python script, import the required modules from the Bokeh 
library.
3)Create a Bokeh plot: Initialize a figure object and customize it according to your 
requirements. Additionally, create a ColumnDataSource object that will hold the data to be 
displayed in the plot.
4)Update the plot data: Define a callback function that updates the data in the ColumnDataSource 
object. This function will be called whenever an event triggers the update.
5)Add interactive elements: Add interactive elements to the plot, such as buttons, sliders, or 
dropdowns, using the available Bokeh models.
6)Register the update function: Use the curdoc().add_periodic_callback() function to register
the update function as a periodic callback. Specify the interval at which the update function
should be called.
7)Create the layout: Arrange the plot and interactive elements into a layout, such as row(), 
column(), or gridplot().
8)Run the Bokeh server: Finally, run the Bokeh server using the bokeh serve command in the
terminal, followed by the name of your Python script.
9)Access the interactive plot: Open a web browser and navigate to the URL provided by the Bokeh 
server (typically http://localhost:5006). You should see your interactive plot that updates in 
real time based on the defined callback function and any user interactions.

Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

Ans-
```
